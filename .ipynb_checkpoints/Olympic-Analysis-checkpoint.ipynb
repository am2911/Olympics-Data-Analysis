```python
import numpy as np
import pandas as pd  
```

```python
df=pd.read_csv("athlete_events.csv")
region_df=pd.read_csv("noc_regions.csv")
```

```python
df.head()
```

```python
region_df.head()
```

```python
# merge with region_df
df=df.merge(region_df,on="NOC",how="left")
```

```python
df.head()
```

```python
# Filtering for summer olympics 
df=df[df["Season"]=="Summer"]
```

```python
df.head()
```

```python
df["region"].shape
```

```python
df.isnull().sum()
```

```python
# Removed Duplicate Values
df.drop_duplicates(inplace=True)
```

```python
df["Medal"].value_counts()
```

```python
# One-hot-encoding medals
df=pd.concat([df,pd.get_dummies(df["Medal"])],axis=1)
df=df.loc[:, ~df.columns[::-1].duplicated()[::-1]]
df.head()
```

```python
years=df["Year"].unique().tolist()
years.sort()
years.insert(0,"Overall")
years
```

```python
country=np.unique(df["region"].dropna().values).tolist()
country.sort()
country.insert(0,"Overal")
country
```

```python
# Grouping medals on the basis of region
df.groupby("region").sum()[["Gold","Silver","Bronze"]].sort_values("Gold",ascending=False).reset_index()
```

```python
# removing duplicate rows to get exact values 
medal_tally=df.drop_duplicates(subset=["Team","NOC","region","Games","Year","City","Sport","Event","Medal"])
```

```python
medal_tally.head()
```

```python
medal_tally=medal_tally.groupby("region").sum()[["Gold","Silver","Bronze"]].sort_values("Gold",ascending=False).reset_index()
medal_tally["Total"]=medal_tally["Gold"]+medal_tally["Silver"]+medal_tally["Bronze"]
medal_tally
```

```python
medal_tally[medal_tally["region"]=="India"]
```

```python
# medal Tally based on years & country
def fetch_medal_tally(df,year,country):
    medal_df=df.drop_duplicates(subset=["Team","NOC","region","Games","Year","City","Sport","Event","Medal"])
    flag=0
    if year=="Overall" and country=="Overall":
        temp_df=medal_df
    
    if year=="Overall" and country!="Overall":
        flag=1
        temp_df=medal_df[medal_df["region"]==country]
    
    if year!="Overall" and country=="Overall":
        temp_df=medal_df[medal_df["Year"]==int(year)]
        
    if year!="Overall" and country!="Overall":
        temp_df=medal_df[(medal_df["Year"]==year) & (medal_df["region"]==country)]
    
    if flag==1:
        x=temp_df.groupby("Year").sum()[["Gold","Silver","Bronze"]].sort_values("Year",ascending=True).reset_index()
    else:
        x=temp_df.groupby("region").sum()[["Gold","Silver","Bronze"]].sort_values("Gold",ascending=False).reset_index()
    x["Total"]=x["Gold"]+x["Silver"]+x["Bronze"]
    print(x)
    
```

```python
fetch_medal_tally(df,year="Overall",country="India")
```
